```python
fixed = [False, True, False, False, True, False]
fid_values = [2.0, 0.2]
```

```python
fiducial_s0 = jsm_mcmc.Chain("../../mcmc/inference_tests/s0/mock_jsm/chain.h5", fixed=fixed, Nstack=500, Nburn=100, Nthin=15)

fiducial_s0_old = jsm_mcmc.Chain("../../mcmc/inference_tests/s0/mock_jsm_old/chain.h5", fixed=fixed, Nstack=500, Nburn=100, Nthin=15)

nadler_s0 = jsm_mcmc.Chain("../../mcmc/inference_tests/s0/mock_nadler/chain.h5", fixed=fixed, Nstack=500, Nburn=100, Nthin=15)
```

```python
binz = np.linspace(0, 18, 50)
plt.hist(nadler_s0.chisq.flatten() - nadler_s0.chisq.flatten().min(), bins=binz, edgecolor="white", density=True, alpha=0.4, color="orange")
plt.hist(fiducial_s0.chisq.flatten() - fiducial_s0.chisq.flatten().min(), bins=binz, edgecolor="white", density=True, alpha=0.4, color="blue")
plt.hist(fiducial_s0_old.chisq.flatten() - fiducial_s0_old.chisq.flatten().min(), bins=binz, edgecolor="white", density=True, alpha=0.4, color="green")
plt.show()
```

```python
chain_labels = ["Danieli's method \n $\\alpha = $" + nadler_s0.constraints[0] + "\n $\\sigma = $" + nadler_s0.constraints[1],
                "New method \n $\\alpha = $" + fiducial_s0.constraints[0] + "\n $\\sigma = $" + fiducial_s0.constraints[1],
                "Old method \n $\\alpha = $" + fiducial_s0_old.constraints[0] + "\n $\\sigma = $" + fiducial_s0_old.constraints[1]]

data_s0 = [nadler_s0.clean, fiducial_s0.clean, fiducial_s0_old.clean]
test = jsm_mcmc.MulitChain(data_s0, chain_labels, fixed)
test.plot_posteriors(paper=True, filledPlots=True, nContourLevels=3, smoothingKernel=1.5, truths=fid_values, colorsOrder=["oranges", "blues", "greens"])#, paramRanges=[[1.9, 2.3], [0, 1]])
```
