### The s0 sample

```python
fixed = [False, True,  True, False, False, False]

s0_1 = jsm_mcmc.Chain("../../mcmc/simple_power_law/s0/alpha_permutations/mock_1/chain.h5", fixed=fixed, Nstack=500, Nburn=100, Nthin=15)

fixed = [False, True, False, False, True, False]

s0_2 = jsm_mcmc.Chain("../../mcmc/simple_power_law/s0/sigma_permutations/mock_1/chain.h5", fixed=fixed, Nstack=500, Nburn=100, Nthin=15)
```
