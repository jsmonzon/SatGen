```python
M = 1e13

m = np.linspace(9e8,9e12,100)

psi = m/M
```

```python
def unevolved(psi):
    gamma = 0.22
    alpha = -0.91
    beta = 6
    omega = 3
    return gamma* np.sign(psi)*(np.abs(psi)) ** (alpha) * np.exp(-beta*(psi)**omega)
```
