```python
truths_S = [10.5, 1.9, 0.2, 0, 0, 0]

plabels = ["$M_{*}$", "$\\alpha$", "$\\sigma$"," $\\gamma$", "$\\beta$", "$\\tau$"]

priors = [[9.5,11.5], [-1,7], [0,5], [-2,2], [-3,2], [-2,2]]

plot_range = [[9.8,11.2], [0,4.2], [0,2.2], [-2,2], [-3,2], [-2,2]]

ndim_short = 3

Nstack = 2000
```

```python
ms_A = jsm_mcmc.single_chain("../../mcmc/model_runs/mock_A/model_A/chain.h5", Nstack, Ndim=ndim_short, truths=truths_S, labels=plabels)

ms_B = jsm_mcmc.single_chain("../../mcmc/model_runs/mock_A/model_B/chain.h5", Nstack, Ndim=ndim_short, truths=truths_S, labels=plabels)

ms_C = jsm_mcmc.single_chain("../../mcmc/model_runs/mock_A/model_C/chain.h5", Nstack, Ndim=ndim_short, truths=truths_S, labels=plabels)

ms_D = jsm_mcmc.single_chain("../../mcmc/model_runs/mock_A/model_D/chain.h5", Nstack, Ndim=ndim_short, truths=truths_S, labels=plabels)
```

```python
input_as = [1.6, 1.8, 2.0, 2.2, 2.4]

data = [ms_A.end[:,1], ms_B.end[:,1], ms_A.end[:,1], ms_C.end[:,1], ms_D.end[:,1]]

# Create a violin plot
plt.figure(figsize=(8, 6))


colors = ['blue', 'green', 'orange', 'red', 'pink']  # Adjust these colors based on your preference

# Create a violin plot with custom x values and colors
plt.figure(figsize=(8, 6))
violins = plt.violinplot(data, showmeans=False, showmedians=True, positions=input_as, widths=0.1, showextrema=False)

# Customize the colors of each violin
for i, violin in enumerate(violins['bodies']):
    violin.set_facecolor(colors[i])

violins['cmedians'].set_color(colors)

plt.xticks(input_as, ['1.6', '1.8', '2.0', '2.2', '2.4'])
plt.ylabel('output a')
plt.xlabel("input a")

xpoints = ypoints = plt.xlim()
plt.plot(xpoints, ypoints, linestyle='--', color='k', scalex=False, scaley=False)

plt.ylim(0, 5)
plt.show()
```
