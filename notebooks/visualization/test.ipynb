```python
import numpy as np
import matplotlib.pyplot as plt

# Define the Gaussian function
def gaussian(x, mean, sigma):
    return np.exp(-0.5 * ((x - mean) / sigma) ** 2)

# Define parameters
mean = 12
sigma = 1

# Generate x values
x = np.linspace(0, 24, 1000)

# Calculate y values
y = gaussian(x, mean, sigma)

# Plot the Gaussian function
plt.axvline(12, ymin=0.05, ymax=0.95, label='$S_{0}}$', color="green")
plt.plot(x, gaussian(x, mean, 0.15), label='$S_{15}}$', color="darkorange")
plt.plot(x, gaussian(x, mean, 0.30), label='$S_{30}}$', color="red")
plt.xlim(10,14)
plt.ylim(-0.05, 1.05)
plt.axhline(0, ls="--", color="grey")
plt.xlabel('Host Halo Mass')
plt.ylabel('Probability Density')
plt.legend(fontsize=15)
plt.savefig("../../../../Downloads/samples_2.pdf", bbox_inches="tight")
plt.show()

```
