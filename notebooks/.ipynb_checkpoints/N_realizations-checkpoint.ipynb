```python
import numpy as np
import scipy
import matplotlib.pyplot as plt

import matplotlib as mpl
    
import sys  
sys.path.insert(0, '../src/')

from astropy.table import Table

from jsm_routines import SHMF, CSMF, SHMR
import os
```

```python
as_data = np.load("../../data/12_4_0/numpy/acc_surv_mass.npy")
as_red = np.load("../../data/12_4_0/numpy/acc_surv_redshift.npy")
```

```python
Ms = SHMR(as_data)

surv_massr, surv_CSMF = CSMF(Ms) # accreation mass of the surviving halos!
```

```python
Nreal = np.array([3000,2000,1000,500,300,100,50,5])


mat = np.zeros(shape=(Nreal.shape[0], 3, 50))
```

```python
for i,N in enumerate(Nreal):

    choice = np.random.choice(as_data.shape[0], size=N)

    surv_massr, surv_CSMF_i = CSMF(Ms[choice], plot=False) # accreation mass of the surviving halos!
    
    mat[i] = surv_CSMF_i 
```

```python
colors = ["blue", "orange", "green", "red", "purple", "brown", "pink", "cyan"]
```

```python
plt.figure(figsize=(10,8))

for i,val in enumerate(mat):
    
    plt.plot(surv_massr, val[1,:], color=colors[i], label="N="+str(Nreal[i]))
    plt.fill_between(surv_massr, y1=val[0, :], y2=val[2, :], alpha=0.1, color=colors[i])

plt.yscale("log")
plt.xscale("log")
plt.xlabel("log m$_{stellar}$ (M$_\odot$)", fontsize=15)
plt.ylabel("N (> m$_{stellar}$)", fontsize=15)
plt.legend()
plt.ylim(0.8)
plt.show()
```

```python
test = mat[5]
```

```python
def scatter_stat(CSMF, det_CSMF):

    if len(CSMF.shape) == 4:
        print("averaging over the samples!")
        CSMF = np.average(CSMF,axis=0)
    
    Nreal = CSMF.shape[0]
    Mpix = CSMF.shape[2]

    stat = (CSMF[:,2]-CSMF[:,0])/(det_CSMF[2]-det_CSMF[0])

    print("returning a", Nreal, "by", Mpix, "matrix")
    return stat
```

```python
stat = scatter_stat(mat,surv_CSMF)
```

```python
plt.figure(figsize=(10,8))

for i,val in enumerate(stat):

    plt.plot(surv_massr, val, color=colors[i], label=str(Nreal[i]))

plt.axhline(1, ls="--", color="black",label="total")
plt.xscale("log")
plt.xlabel("log m$_{stellar}$ (M$_\odot$)", fontsize=15)
plt.ylabel("$ (N_{95}-N_5) / (|N_{95}-N_5|) $")
plt.legend()
plt.show()
```
