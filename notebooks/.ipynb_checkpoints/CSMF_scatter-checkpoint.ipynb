```python
import numpy as np
import scipy
import matplotlib.pyplot as plt

import matplotlib.cm as cm
    
import sys  
sys.path.insert(0, '../src/')

from astropy.table import Table

from jsm_routines import SHMF, CSMF
import os
```

```python
surv_data = np.load("../../data/12_4_0/numpy/surv_mass.npy")
```

```python
surv_massr, surv_CSMF = CSMF(surv_data) # determinisitic
```

```python
surv_massr_015, surv_CSMF_015 = CSMF(surv_data, scatter=0.25) # determinisitic
```

```python
surv_massr_5, surv_CSMF_5 = CSMF(surv_data, scatter=0.5) # determinisitic
```

```python
surv_massr_10, surv_CSMF_10 = CSMF(surv_data, scatter=1) # determinisitic
```

```python
surv_massr_15, surv_CSMF_15 = CSMF(surv_data, scatter=1.5) # determinisitic
```

```python
plt.figure(figsize=(8, 8))

plt.title("Increasing scatter in SHMR")


plt.plot(surv_massr, surv_CSMF[1, :], label="Deterministic", color="red")
plt.fill_between(surv_massr, y1=surv_CSMF[0, :], y2=surv_CSMF[2, :], alpha=0.2, color="red")

plt.plot(surv_massr_015, surv_CSMF_015[0, :], color="blue", ls=":", label="0.25 dex scatter")
plt.plot(surv_massr_015, surv_CSMF_015[2, :], color="blue", ls=":")

plt.plot(surv_massr_5, surv_CSMF_5[0, :], color="green", ls=":", label="0.5 dex scatter")
plt.plot(surv_massr_5, surv_CSMF_5[2, :], color="green", ls=":")

plt.plot(surv_massr_10, surv_CSMF_10[0, :], color="orange", ls=":", label="1.0 dex scatter")
plt.plot(surv_massr_10, surv_CSMF_10[2, :], color="orange", ls=":")

plt.plot(surv_massr_15, surv_CSMF_15[0, :], color="black", ls=":", label="1.5 dex scatter")
plt.plot(surv_massr_15, surv_CSMF_15[2, :], color="black", ls=":")





plt.yscale("log", base=10)

plt.xlabel("m$_{stellar}$ (M$_\odot$)", fontsize=15)
plt.ylabel("N (> m$_{stellar}$)", fontsize=15)
plt.xlim(3.05)
plt.legend(fontsize=12)
plt.grid(alpha=0.2)
plt.savefig("../figures/CSMF_scatter.png")

plt.show()
```
