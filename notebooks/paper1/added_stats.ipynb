```python
plt.style.use('../paper/paper.mplstyle')
double_textwidth = 7.0 #inches
single_textwidth = 3.5 #inches
```

```python
fixed = [False, True, False, False, True, False]

fid_values = [2.0, 0.2]

fiducial_s0 = jsm_mcmc.Chain("../../mcmc/inference_tests/added_stats/added_stats_s0/mock_og/chain.h5", fixed=fixed, Nstack=500, Nburn=100, Nthin=15)
total_s0 = jsm_mcmc.Chain("../../mcmc/inference_tests/added_stats/added_stats_s0/mock_tot/chain.h5", fixed=fixed, Nstack=500, Nburn=1000, Nthin=55)

fiducial_s15 = jsm_mcmc.Chain("../../mcmc/inference_tests/added_stats/added_stats_s15/mock_og/chain.h5", fixed=fixed, Nstack=500, Nburn=100, Nthin=15)
total_s15 = jsm_mcmc.Chain("../../mcmc/inference_tests/added_stats/added_stats_s15/mock_tot/chain.h5", fixed=fixed, Nstack=500, Nburn=1000, Nthin=55)

fiducial_s30 = jsm_mcmc.Chain("../../mcmc/inference_tests/added_stats/added_stats_s30/mock_og/chain.h5", fixed=fixed, Nstack=500, Nburn=100, Nthin=15)
total_s30 = jsm_mcmc.Chain("../../mcmc/inference_tests/added_stats/added_stats_s30/mock_tot/chain.h5", fixed=fixed, Nstack=500, Nburn=1000, Nthin=55)
```

```python
fig, axes = plt.subplots(1, 3, sharex=True, sharey=True, figsize=(double_textwidth, single_textwidth))

sns.kdeplot(x=fiducial_s0.clean[:, 0], y=fiducial_s0.clean[:, 1], ax=axes[0], levels=[1-0.99, 1-0.95, 1-0.68], bw_adjust=2, color="grey")
sns.kdeplot(x=total_s0.clean[:, 0], y=total_s0.clean[:, 1], ax=axes[0], levels=[1-0.99, 1-0.95, 1-0.68], bw_adjust=2, color="green")
axes[0].legend(handles=[Line2D([], [], color='green', lw = 2), Line2D([], [], color='grey', lw = 2)], labels=["$M_{*}^{\mathrm{tot}}, S_{0}$", "$M_{*}^{\mathrm{max}}$"], fontsize=11, loc=2, framealpha=1)

sns.kdeplot(x=fiducial_s15.clean[:, 0], y=fiducial_s15.clean[:, 1], ax=axes[1], levels=[1-0.99, 1-0.95, 1-0.68], bw_adjust=2, color="grey")
sns.kdeplot(x=total_s15.clean[:, 0], y=total_s15.clean[:, 1], ax=axes[1], levels=[1-0.99, 1-0.95, 1-0.68], bw_adjust=2, color="darkorange")
axes[1].legend(handles=[Line2D([], [], color='darkorange', lw = 2), Line2D([], [], color='grey', lw = 2)], labels=["$M_{*}^{\mathrm{tot}}, S_{15}$", "$M_{*}^{\mathrm{max}}$"], fontsize=11, loc=2, framealpha=1)

sns.kdeplot(x=fiducial_s30.clean[:, 0], y=fiducial_s30.clean[:, 1], ax=axes[2], levels=[1-0.99, 1-0.95, 1-0.68], bw_adjust=2, color="grey")
sns.kdeplot(x=total_s30.clean[:, 0], y=total_s30.clean[:, 1], ax=axes[2], levels=[1-0.99, 1-0.95, 1-0.68], bw_adjust=2, color="red")
axes[2].legend(handles=[Line2D([], [], color='red', lw = 2), Line2D([], [], color='grey', lw = 2)], labels=["$M_{*}^{\mathrm{tot}}, S_{30}$", "$M_{*}^{\mathrm{max}}$"], fontsize=11, loc=2, framealpha=1)

for ax in axes.flatten():
    ax.axhline(0.2, ls="--", color="grey")
    ax.axvline(2.0, ls="--", color="grey")
    ax.set_xlabel("$\\alpha$")
    ax.set_ylabel("$\sigma$")
    ax.set_ylim(0)

plt.tight_layout()
plt.savefig("../../../paper_1/figures/aux/added_stats.pdf", bbox_inches="tight")
plt.show()
```
